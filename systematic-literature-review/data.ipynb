```python
import pandas as pd
import matplotlib.pyplot as plt

PUBLICATION_YEAR = 'Publication Year'
SYNONYMUS = 'Synonymus / Alike Terms'
TERM = 'Term'
PICOC = 'PICOC'
TITLE = 'Title'
CITED_BY = 'Cited by'
ITEM_TYPE = 'Item Type'

plt.rcParams["figure.autolayout"] = True

df = pd.read_csv('data/articles_with_cites.csv')

df = df.groupby([PUBLICATION_YEAR])[PUBLICATION_YEAR].count().sort_index()

plt.bar(df.index, df.values,color='green')
plt.plot(df.index, df.values, marker='*', color='green')

plt.show()
```
